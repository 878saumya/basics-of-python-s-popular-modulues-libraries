```python
from matplotlib import pyplot as plt
from matplotlib import style 
style.use('ggplot')
x=[5,6,7]
y=[8,5,9]
x1=[3,6,7]
y1=[6,9,1]
x2=[5,6,4]
y2=[2,5,7]
plt.plot(x,y,'g',linewidth=5)
plt.plot(x1,y1,'c', linewidth=5)
plt.plot(x2,y2,'r',linewidth=5)
plt.title('info_test')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.legend('x x1 x2')
plt.show()
```

```python
import matplotlib.pyplot as plt
slices =[7,5,6,8]
activity=["sleep","eat","work","play"]
plt.pie(slices,
        labels=activity,
        startangle=90,
      explode=(0,0,0.1,0)
         )
plt.title('time_spend')
plt.show()

```
